## Dealing with Missing Values

#### What is missing values:

Missing values occur when no data value is stored for variable (feature) in an observation.
Can be represented as “?” , “N/A”, 0, or just a blank cell.

#### How to resolve missing values?

* Check with the data collection source to replace missing value?
* Drop Missing Values (either drop whole variable or just drop single value/row)
* Replace the missing values 
        * Replace with the average value of the entire column or of similar data point).
        * Replace it with frequency/mode value
        * Replace it based on other functions/conditions
        * Sometime leave it missing

### Treating Missing Values of a Single Column (Series)

```python
df_age=df[['Age']]
df_age.head()
```

```python
df_age.isnull().sum()
```

```python
#Method # 1: replacing with some value such as zero
df_age['Age']=df_age['Age'].fillna(0)
df_age.isnull().sum()
```

```python
#Method # 2: replacing with mean value
df_age1=df[['Age']]
df_age1.isnull().sum()
```

```python
a=df_age1['Age'].mean()
df_age1['Age']=df_age1['Age'].fillna(a)

#or 

df_age1['Age']=df_age1['Age'].fillna(df_age1['Age'].mean())
df_age.isnull().sum()
```

```python
#filled missing age with the median age of individuals
df['Age']=df['Age'].fillna(df['Age'].median())
df.isnull().sum()
```

```python
#filled missing Embarked - the catagorical value with some value
df['Embarked']=df['Embarked'].fillna('T')
df.isnull().sum()
```

Data is being normalized for making the values consistent and uniform so that they may have similar/consistent impact on the model. Hence, features can be compared with each other.

## Example # 1 (Single feature scaling)

```python
((df['Fee']-df['Fee'].min())/(df['Fee'].max()-df['Fee'].min())*100).hist()
```

```python
((df['Fee']-df['Fee'].mean())/df['Fee'].std()).hist()
```

## Creating dummy/quantities variables from the categorical variables (one hot encoding)
Most statistical models don’t take the categorical/string/object variables as input hence; categorical variables needs to be converted into the dummy variables as input.
### Solution:
* Add dummy variables for each unique category
* Assign 0 or 1 to each category
* This is called one hot encoding

```python
#Creates dummies into a new dataframe
dummies=pd.get_dummies(df['Sex'])
dummies.head()
```

```python
df.head()
```

```python
#inserting dummy variables in the main dataframe
df = pd.concat([df, dummies], axis=1)
#df = df.drop(['Sex'], inplace=True, axis=1)
```

* To increase the accuracy of the model
* To have better understanding with the distribution of the model

### Examples of binning

Converting age into age groups 
    * Low age group (under 29 years)
    * Middle age group (30-50 years)
    * High Age group (50+ years) 

Converting income into income quintiles 
    * Q1 (Less the 10k)
    * Q2 (10k-20k)
    * Q3 (20k-30K)
    * Q4 (30K-40K)
    * Q5 (40k-50k)

### Creating Binning Automatically using linspace function

```python
df['Age'].plot(kind='hist')
```

```python
bins = np.linspace(min(df['Age']), max(df['Age']),4)
bins
```

```python
Group_name=['Low','Medium','High']
df['Age-bin']=pd.cut(df['Age'], bins, labels=Group_name, include_lowest=True)
```
